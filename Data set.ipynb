```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of rows
n = 1000

# Generate synthetic loan dataset
data = {
    'Loan_ID': [f'LN{i:05d}' for i in range(n)],
    'Gender': np.random.choice(['Male', 'Female', np.nan], n, p=[0.45, 0.45, 0.10]),
    'Married': np.random.choice(['Yes', 'No', np.nan], n, p=[0.6, 0.35, 0.05]),
    'Dependents': np.random.choice(['0', '1', '2', '3+', np.nan], n),
    'Education': np.random.choice(['Graduate', 'Not Graduate'], n),
    'Self_Employed': np.random.choice(['Yes', 'No', np.nan], n, p=[0.1, 0.8, 0.1]),
    'ApplicantIncome': np.random.randint(1500, 25000, n),
    'CoapplicantIncome': np.round(np.random.uniform(0, 10000, n), 2),
    'LoanAmount': np.where(np.random.rand(n) < 0.1, np.nan, np.random.randint(50, 700, n)),
    'Loan_Amount_Term': np.random.choice([360.0, 180.0, 120.0, np.nan], n, p=[0.8, 0.1, 0.05, 0.05]),
    'Credit_History': np.random.choice([1.0, 0.0, np.nan], n, p=[0.85, 0.10, 0.05]),
    'Property_Area': np.random.choice(['Urban', 'Rural', 'Semiurban'], n),
    'Loan_Status': np.random.choice(['Y', 'N'], n, p=[0.7, 0.3])  # This is the output/target
}

# Create DataFrame
df = pd.DataFrame(data)

# Output: Show first 10 rows
print(df.head(10))
```
